# Objetivo 

Hacer el conteo de fotones de una fuente luminosa para un tiempo de coteo determinado. Y con este definir el tipo de fuente luminosa a través de su estadística y su función de correlación de segundo orden $g^{(2)}(0)$.

## I Clasificación de fuentes 

### Fuente térmica 

Es aquella cuya radiación electromagnética es producida por un cuerpo en función de su temperatura, y su función de probabilidad esta dada por:

$$P_{\omega,n} = \frac{\bar{n}^n}{(\bar{n} + 1)^{n+1}} $$

$$(<\Delta\hat{n}>)^2 = <\hat{n}^2> - <\hat{n}>^2  $$

Es la estadística de Bose-Einstein.

### Fuente coherente

Es aquella cuyas ondas electromagnéticas que emiten tienen una fase, amplitud y frecuencia angular constante. Y su función de probabilidad esta dada por:

$$ P_n = \frac{\bar{n}^n e^{-\bar{n}}}{n!} $$

$$(<\Delta\hat{n}>)^2 = \hat{n}$$

Es la estadística de Poisson.

### Fuente cuántica 

Es aquella que emite un número constante de fotones en un intervalo de tiempo determinado, y su estadística esta dada por:

$$P_n = |<n|m>|^2 = \delta_{nm}$$

idealmente.

## II Cristales BBO

Bario Borato, es un compuesto inorgánico, un borato de bario con formula química $BaB_2O_4$. El cristal existe a altas temperaturas, fase $\alpha$ y a bajas temperaturas, fase $\beta$, abreviado como *BBO*, en ambas fases es birrefrigente, y comunmente es un material óptico no-lineal.

Cuando un fotón de frecuencia $\omega_0$ incide sobre un medio dieléctrico, se generan dos fotones con frecuencias menores $\omega_1$ y $\omega_2$.

## III Función de correlación 

Dado que le histograma no es siempre preciso para determinar el tipo de fuente luminosa, es utilizado otro método a través de cual se obtiene la función de correlación de segundo orden.

$$g^{(2)}(\tau) = \frac{<I(t+\tau)I(t)>}{<I(t+\tau><I(t)>} $$

Representa la probabilidad de detectar, en una posición dada, un fotón en $t$ y también otro en $t+\tau$, esto es, con un retardo $\tau$.

Si $g^{(2)}(\tau) = 1$, entonces los fotones van llegando de forma independiente

### Función de correlación de segundo orden clásica 

$$g^{(2)}(\tau) = 1 + e^{\frac{-2\tau}{t_0}}$$

Los fotones tienden a aparecer juntos. La probabilidad de que se produzca la detección de un segundo fotón después de un retardo $\tau$ decrece con $\tau$. Se trata de un efecto clásico.

### Función de correlación de segundo orden cuántica 

$$g^{(2)}(\tau) = 1 - e^{(-r+\gamma)\tau} < 1$$

donde la $r$ es la tasa de bombeo y $\gamma^{-1}$ es el tiempo de vida del estado excitado. 

La probabilidad de que se produzca detecciones e coincidencia de fotones en un intervalo $\tau$ es menor que en el caso de estado coherente.
